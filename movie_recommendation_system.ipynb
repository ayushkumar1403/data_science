```python
# important line of code
# first we import the json function
# then start the function
# then we devclare l in which we will later store the values
# iterating  over the jason loads obj
# and then storing the values in l and then in last returing the values
import json
def convert (obj):
  l = []
  for i in json.loads(obj):
    l.append(i['name'])
  return l
```

```python
def convertcast(obj):
  l = []
  count = 0
  for i in json.loads(obj):
    if count < 3:
      l.append(i['name'])
    count = count +1
  return l
```

```python
def fetch_director (obj):
  l = []
  for i in json.loads(obj):
    if i['job']=='Director':
      l.append(i['name'])
      break
  return l
```

```python
movies_2['crew']= movies_2['crew'].apply(fetch_director)
```

```python
# as all the other columns are lists and the overview column is a string so we will first convert the string to list by splitting them and putting a comma between them
movies_2["overview"] = movies_2["overview"].apply( lambda x : x.split())
```

```python
movies_2["tags"] = movies_2["genres"] + movies_2["keywords"] + movies_2["crew"] + movies_2["cast"] + movies_2["overview"]
```

```python
movies = movies_2.drop(columns= ["overview","cast","keywords","crew","genres"])
```

```python
# to remove the comma between them and converting the list to string using the below code
# it takes a list of strings and create a single string
# we have wrote " ".join(x) because we want to join the stringh and connect it with spaces in between
movies["tags"] = movies["tags"].apply(lambda x : " ".join(x))
```

```python
movies
```

```python
# now we will convert all the tags to lowercase because it is good to have practice
# lambda is a smaller way to use functions
movies["tags"] = movies["tags"].apply( lambda x : x.lower())
```

```python
movies
```

```python
# now to give a unique identity to the data we will convert the text to text vector that we will later use to find similarity between
# any two movies and to recommend the most similar movies
# it is a part of natural language processing
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features = 5000,stop_words="english")
```

```python
vector_values = cv.fit_transform(movies["tags"]).toarray()
```

```python
data = cv.get_feature_names_out()
```

```python
# learn how to see the whole dataset for the data below
print(data)
```

```python
# now to remove similar words like actor and actors because they have same meaning
# this part no need to understand because it is a part of nlp which you have not studied
import nltk
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python

# this part dont need to understand because it is a part of natural language processing
def stem (text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
movies['tags'] = movies['tags'].apply(stem)
```

```python
# to find simailarity between movies between each other
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector_values)
```

```python

similarity
```

```python
# enumerate(movies) is used to iterate over the list movies along with its index
# list ennumerate function is used for sorting by taking the index value also
# reverse is true because we want to take the values in decreasing order and we usd lambda fuction to say that the
#  sorting should be done for the simarity and not for the index
def recommend(movie):
  movie_index = movies[movies["title"] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse = True, key = lambda x:x[1])[1:6]

  for i in movies_list:
    print(movies.iloc[i[0]].title)
```

```python
recommend("Batman Begins")
```
